```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtering the data for damage and no-damage cases
damage_cases = gdf[gdf['Damage'] == 1]['slope20']
no_damage_cases = gdf[gdf['Damage'] == 0]['slope20']

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plotting histograms
plt.subplot(1, 2, 1)
plt.hist(damage_cases, bins=30, alpha=0.5, label='Damage = 1')
plt.hist(no_damage_cases, bins=30, alpha=0.5, label='Damage = 0')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.title('Histogram of Slope for Damage and No-Damage Cases')
plt.legend()

# Plotting density plots
plt.subplot(1, 2, 2)
sns.kdeplot(damage_cases, shade=True, label='Damage = 1')
sns.kdeplot(no_damage_cases, shade=True, label='Damage = 0')
plt.xlabel('Slope')
plt.ylabel('Density')
plt.title('Density Plot of Slope for Damage and No-Damage Cases')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Define the intervals for building value
intervals = pd.interval_range(start=0, end=int(max(gdf['b_div_c'])) + 500000, freq=500000)

# Categorize 'b_div_c' values into defined intervals
gdf['b_div_c_interval'] = pd.cut(gdf['b_div_c'], intervals)

# Calculate counts for each interval
interval_counts = gdf['b_div_c_interval'].value_counts().sort_index()

# Set pandas option to display all rows
pd.set_option('display.max_rows', None)

# Print the counts for each interval
print(interval_counts)

# Reset pandas option to its default (optional)
pd.reset_option('display.max_rows')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Function to format the x-axis labels
def millions_formatter(x, pos):
    return f'{int(x / 1e6)}M'

# Filtering the data for damage and no-damage cases
damage_cases = gdf[gdf['Damage'] == 1]['b_div_c']
no_damage_cases = gdf[gdf['Damage'] == 0]['b_div_c']

# Define bin edges for histogram with a 100,000 interval, up to 20,000,000
bins = np.arange(0, 20000000, 100000)

# Plot histogram
plt.figure(figsize=(10, 6))

plt.hist(damage_cases, bins=bins, alpha=0.5, label='Damage', density=True, color='red')
plt.hist(no_damage_cases, bins=bins, alpha=0.5, label='No Damage', density=True, color='blue')

plt.xlabel('Building value DKK')
plt.ylabel('Percentage')
plt.title('Building Value')
plt.legend()

# Format x-axis labels
formatter = FuncFormatter(millions_formatter)
plt.gca().xaxis.set_major_formatter(formatter)

plt.show()
```
